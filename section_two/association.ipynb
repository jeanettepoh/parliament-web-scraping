```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_file = os.path.join(os.getcwd(), "data", "residential_data.csv")
df = pd.read_csv(data_file)
df['transaction_date'] = pd.to_datetime(df['transaction_date'])
df['Year'] = df['transaction_date'].dt.year
```

```python
df
```

```python
df = df[(df["represented"] == "SELLER") & (df["transaction_type"] == "RESALE")]
df
```

```python
df = df[df["salesperson_name"] != "-"]
df
```

```python
df["salesperson_name"].value_counts()
```

```python
df["Year"].value_counts()
```

```python
# Directory to save figures
save_dir = "figures"
os.makedirs(save_dir, exist_ok=True)
```

```python
# Get unique years from the Year column
unique_years = df['Year'].unique()

# Loop over unique years
for year in unique_years:
    # Filter data for the current year
    df_filtered = df[df['Year'] == year]
    
    # Count number of rows (transactions) for each agent
    agent_counts = df_filtered['salesperson_name'].value_counts()
    
    # Plotting the distribution of number of transactions by agent for the current year
    plt.figure(figsize=(8, 5))
    plt.bar(agent_counts.index, agent_counts.values, color='red')
    plt.title(f'Distribution for Number of Sales in {year}')
    plt.xlabel('Agent')
    plt.ylabel('Number of Transactions')
    plt.xticks([])  # Remove x-axis ticks
    # Set y-axis limit to max value + 10
    print(agent_counts.max())
    plt.ylim(0, agent_counts.max() + 10)
    plt.tight_layout()
    plt.savefig(os.path.join(save_dir, f"Sales_Distribution_for_{year}.png"))
    plt.show()
```
